```python
# Clean club names in both dataframes
df['Club'] = df['Club'].str.strip().str.lower()
prdf['Club'] = prdf['Club'].str.strip().str.lower()
```

```python
df['Previous'] = df.apply(
    lambda row: prdf.loc[
        (prdf['Club'] == row['Club']) & (prdf['Year'] == row['Year'] - 1), '#'
    ].values[0] if not prdf.loc[
        (prdf['Club'] == row['Club']) & (prdf['Year'] == row['Year'] - 1), '#'
    ].empty else 'New',
    axis=1
)
missing = df[~df['Club'].isin(prdf['Club'])]['Club'].unique()
print("Clubs not found in previous data:", missing)

```

```python
df= df.sort_values(by='Year', ascending=False)
```

```python
from sklearn.model_selection import train_test_split
x=df[['Previous','previous_point','Transfer']]
y=df['#']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

x_train.describe()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import GradientBoostingRegressor

model = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('regression', GradientBoostingRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=4,
        random_state=42
    ))
])
model.fit(x_train, y_train)
```

```python


y_pred = model.predict(x_test)

```

```python
y_pred= y_pred.astype(int)
y_pred
```

```python
from sklearn.metrics import  mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)

r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")

print(f"R² Score: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))


plt.plot(y_test.values, label='Actual', marker='o')


plt.plot(y_pred, label='Predicted', marker='x')

plt.title("Actual vs Predicted Transfer Points")
plt.xlabel("Sample Index")
plt.ylabel("Points")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
